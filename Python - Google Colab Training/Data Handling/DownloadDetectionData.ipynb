```
trainPercent  = 10
pathTrain = '/content/drive/My Drive/Data/Detection/Training/'
with open('/content/drive/My Drive/Data/Detection/Training/FaultRegister.csv','r') as f:
    faultRegister = np.genfromtxt(f,delimiter = ",")
sortedReg = faultRegister[faultRegister[:,2].argsort()]
nFiles = int(trainPercent /100/2 * faultRegister[-1,0])
fileIndices = np.concatenate([sortedReg[:nFiles,0].astype(int),sortedReg[-nFiles:,0].astype(int)])
listdirTrain = []
destination = '/content/drive/My Drive/Download'
for index in fileIndices:
    listdirTrain.append('Train_' + str(index) + '.tfrecord')
```

```
for fileName in listdirTrain:
    if not os.path.isfile(destination+fileName):
        print('Copying file: ' + fileName)
        shutil.copyfile(pathTrain+fileName, destination+fileName)
    
```
