```
# Testing

path = '/content/drive/My Drive/DataRaw/Detection/Testing/'
openPattern = path + 'DataOpenFault4N_*.csv'
closedPattern = path + 'DataClosedFault4N_*.csv'
openPattern2 = path + 'DataOpenFault4N_*.csv'
closedPattern2 = path + 'DataClosedFault4N_*.csv'
globPatterns = [openPattern,closedPattern,openPattern2,closedPattern2]
openPatterns = [openPattern,openPattern2]
closedPatterns = [closedPattern,closedPattern2]
# Seed is 6 DO NOT CHANGE
seed = 6

# Define model parameters
# Create Dataset
closedList =  tf.io.gfile.glob(closedPatterns)
openList = tf.io.gfile.glob(openPatterns)
```

```

filePathWrite = '/content/drive/My Drive/Data/Detection/Testing/'
closedIntList = []
for i,fileName in enumerate(closedList):
    with open(fileName,'r') as f:
        data = np.genfromtxt(f,delimiter = ",")
        fInt = data[0,4]

    closedIntList.append((fileName,fInt))

    print(100*i/len(closedList))
closedIntList.sort(key=lambda x:x[1])
```

```

with open(filePathWrite + 'closedFaultIntensityList_Sorted.csv','w') as f:
    for fileName,intensity in closedIntList:
        update = fileName + ',' + str(intensity) + '\n'
        f.write(update)
```

```
filePathWrite = '/content/drive/My Drive/Data/Detection/Testing/'
openIntList = []
for i,fileName in enumerate(openList):
    with open(fileName,'r') as f:
        data = np.genfromtxt(f,delimiter = ",")
        fInt = data[0,4]

    openIntList.append((fileName,fInt))

    print(100*i/len(openList))
openIntList.sort(key=lambda x:x[1])
```
